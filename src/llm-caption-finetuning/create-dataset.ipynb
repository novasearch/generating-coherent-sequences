{ Instruction<sub>1</sub>, ..., Instruction<sub>n</sub> ; ImageCaption<sub>n</sub> }

```python
import json
from tqdm import tqdm

RECIPES_PATH = '.json'
IMAGES_PATH = ''
CAPTIONS_FILE = '.json'
RESULT_FILE = 'blip_caption_dataset_v2.json'

CAPTION_TYPE = "context"

with open(RECIPES_PATH, 'r') as f:
    recipes = json.load(f)

with open(CAPTIONS_FILE, 'r') as f:
    captions = json.load(f)

results = []

recipe_items = recipes.items()

for recipe_id, _recipe in tqdm(recipe_items):
    try:
        recipe = _recipe['recipe']

        previous_steps = ""

        for step in recipe['instructions']:
            previous_steps += f"Step {step['stepNumber']}: {step['stepText']}\n"
            image_id = f"{recipe_id}_{step['stepNumber']}"
            caption = captions[image_id][CAPTION_TYPE]

            sample = {
                "image": image_id,
                "steps": previous_steps,
                "caption": caption
            }

            results.append(sample) 
    except KeyError as ke:
        print(ke)
        continue

with open(RESULT_FILE, "w") as f:
    json.dump(results, f, indent=4)
```

{ Caption<sub>1</sub>, ..., Caption<sub>n-1</sub>, Instruction<sub>n</sub> ; ImageCaption<sub>n</sub> }

```python
import json
from tqdm import tqdm

RECIPES_PATH = '.json'
IMAGES_PATH = ''
CAPTIONS_FILE = '.json'
RESULT_FILE = 'blip_instructions_and_prompt_to_caption_dataset_v2.json'

CAPTION_TYPE = "context"

with open(RECIPES_PATH, 'r') as f:
    recipes = json.load(f)

with open(CAPTIONS_FILE, 'r') as f:
    captions = json.load(f)

results = []

recipe_items = recipes.items()

for recipe_id, _recipe in tqdm(recipe_items):
    try:
        recipe = _recipe['recipe']

        previous_steps = ""

        for step in recipe['instructions']:
            image_id = f"{recipe_id}_{step['stepNumber']}"
            caption = captions[image_id][CAPTION_TYPE]
            current_step_input = previous_steps + f"Step {step['stepNumber']}: {step['stepText']}\n" 

            previous_steps += f"Step {step['stepNumber']}: {caption}\n"

            sample = {
                "image": image_id,
                "steps": current_step_input,
                "caption": caption
            }

            results.append(sample) 
    except KeyError as ke:
        print(ke)
        continue

with open(RESULT_FILE, "w") as f:
    json.dump(results, f, indent=4)
```

Alpaca format

```python
import json
from sklearn.model_selection import train_test_split

INPUT = '.json'
ALPACA_DATA = ".json"

INSTRUCTION = "Give me the image caption for this recipe."

dataset = []

with open(INPUT, "r") as f:
    input_dataset = json.load(f)

for sample in input_dataset:
    sample_alpaca = {
        "instruction": INSTRUCTION,
        "input": sample["steps"],
        "output": sample["caption"]
    }

    dataset.append(sample_alpaca)

train, eval = train_test_split(dataset, test_size=0.2)

with open(ALPACA_DATA.format("train"), "w") as f:
    json.dump(train, f, indent=4)

if eval:
    with open(ALPACA_DATA.format("eval"), "w") as f:
        json.dump(eval, f, indent=4)

print(f"Dataset has {len(dataset)} examples.")
```

Alpaca format but all in instruction

```python
import json


INPUT = '.json'
ALPACA_DATA = '.json'

INSTRUCTION = "Give me the image caption for this recipe."

dataset = []

with open(INPUT, "r") as f:
    input_dataset = json.load(f)

for sample in input_dataset:
    sample_alpaca = {
        "instruction": f"{INSTRUCTION}\n{sample['steps']}",
        "input": "",
        "output": sample["caption"]
    }

    dataset.append(sample_alpaca)

with open(ALPACA_DATA, "w") as f:
    json.dump(dataset, f, indent=4)

print(f"Dataset has {len(dataset)} examples.")
```

```python
ALPACA_DATA_TRAIN = '.json'
ALPACA_DATA_VAL = '.json'

NUM_EXAMPLES=5000

with open(ALPACA_DATA_TRAIN, "w") as f:
    json.dump(dataset[:NUM_EXAMPLES-500], f, indent=4)

with open(ALPACA_DATA_VAL, "w") as f:
    json.dump(dataset[NUM_EXAMPLES-500:], f, indent=4)
```
